# Forecasting

```python
import os
import pandas as pd
import sqlalchemy
import ydata_profiling
```

## DB Connection

```python
user = os.getenv("POSTGRES_USER")
password = os.getenv("POSTGRES_PASSWORD")
server = os.getenv("POSTGRES_SERVER")
port = os.getenv("POSTGRES_PORT")
db = os.getenv("POSTGRES_DB")

database_uri = f"postgresql://{user}:{password}@{server}:{port}/{db}"

engine = sqlalchemy.create_engine(database_uri)
```

## Data

```python
query = lambda balance_id: f"""
SELECT
  SUM(balance_movement_amount) AS balance_movement_amount,
  balance_id,
  balance_movement_date
FROM balance_movements
WHERE balance_id = {balance_id}
GROUP BY
  balance_movement_date,
  balance_id
ORDER BY
  balance_movement_date ASC
;
"""
```

```python
pd.read_sql(query(120), con=engine)
```
